```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
# Set the default figure size for the notebook, so they are not all tiny
matplotlib.rcParams['figure.figsize'] = (8,6)
%config InlineBackend.figure_formats = ['svg']

# Colors that can be read on a projector
matplotlib.rcParams['image.cmap'] = 'Set1'
```

Here's a univariate dataset that will be tricky to classify. One class surrounds the other:

```python
np.random.seed(1)
X = np.random.uniform(low=-10, high=10, size=100)
y = np.ones_like(X, dtype=np.int64)
y[X > 3] = -1
y[X < -3] = -1
```

Here's what that looks like in X:

```python
plt.scatter(X, y=np.zeros_like(X), c=y, alpha=0.5)
plt.ylim(-10, 10)
plt.xlabel("X")
plt.ylabel("fake dimension for plotting")
plt.show()
```

The classes are not separable -- at least in *X*. But they are separable if you expand to two dimensions by adding an $X^2$ term!

```python
X = np.column_stack([X, X**2])
```

```python
plt.scatter(X[:,0], y=X[:,1], c=y, alpha=0.5)
plt.ylim(0, 100)
plt.xlabel("X")
plt.ylabel("X^2")
plt.show()
```

If we fit an SVM to the expanded data with a *linear* kernel, we can find a good boundary, even if we couldn't do so on the original data:

```python
from sklearn import svm
```

```python
svm_fit = svm.SVC(C=1, kernel='linear').fit(X, y)
svm_fit
```

```python
# Plotting decision regions in python is pretty annoying.
# The mlxtend package seems to help!
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X=X, y=y, clf=svm_fit, legend=0)
plt.show()
```
